```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df_ad_expenditure = pd.read_csv("scatter_plot_ii.csv")
df_ad_expenditure
```

```python
#plt.figure(figsize=(10,8))
sns.set(rc = {'figure.figsize': (9,6)})
sns.regplot(x = "Budget",
            y = "Sales",
            data = df_ad_expenditure,
            scatter_kws = {'color' :'k'},
            line_kws = {'color' : 'red'})
plt.xlabel("Ad Expenditure in (000's of $) ", weight="bold")
plt.ylabel("Sales in (000's of Units)" , weight="bold")
plt.title("Effect of Ad Expenditure on Sales", fontsize=14 , fontweight="bold")
plt.show()
```

```python
sns.lmplot(x = "Budget",
            y = "Sales",
            data = df_ad_expenditure,
            height = 10,
            scatter_kws = {'color' :'k'},
            line_kws = {'color' : 'red'})
plt.xlabel("Ad Expenditure in (000's of $) ", weight="bold")
plt.ylabel("Sales in (000's of Units)" , weight="bold")
plt.title("Effect of Ad Expenditure on Sales", fontsize=14 , fontweight="bold")
plt.show()
```

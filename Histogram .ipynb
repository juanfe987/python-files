```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df_real_estate = pd.read_csv("histogram_data.csv")
df_real_estate
```

```python
sns.set_style("white")
plt.figure(figsize=(8,6))
plt.hist(df_real_estate["Price"],
        bins =8,
        color = "#108A99")
plt.title("Distribution of Real Estate Prices", fontsize=14 , fontweight="bold")
plt.xlabel("Price in (000' $)",fontsize=13)
plt.ylabel("Numer of Properties",fontsize=13)
sns.despine()
plt.show()
```
